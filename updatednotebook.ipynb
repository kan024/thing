# Research Question

Is there a relationship between college student's amount of physical activity in terms of average time spent exercising and walking daily and their mental health measured through self-reported positive affect and stress levels to indirectly investigate the effect of access to urban green spaces on health?

## Background and Prior Work


The effects of urban green spaces on public health have been of great interest in recent years. Green spaces such as parks and gardens facilitate forms of physical activity, mental rejuvenation, and relief from stress, leading to direct and indirect advantages for people’s mental and physical health. This eco-friendly solution is important because while individual people can be positively affected, the environment as a whole can also benefit from them, improving qualities such as climate change, biodiversity, and air quality. Because of this, recently worldwide efforts have been made to provide more urban green spaces in order to create more sustainable and comfortable living environments. In our project, we want to focus on the connection between urban green spaces and college students, specifically how these areas have an effect on their mental health.

One study done within this field explored the relationship between urban green spaces and public health by doing a literature search on other studies and reviews of the different health effects of green spaces. <a name="cite_ref-1"></a>[<sup>1</sup>](#cite_note-1) The articles that were found were based on their relevance, then graded and reviewed accordingly using a scale. The result of this study was that there is very weak evidence for the relationship between physical and mental health and urban green spaces. They also found that green spaces have an overall positive effect on general health. While this study does a good job of exploring the relationship between urban green spaces and overall health, we believe that it is not an entirely accurate representation, as not all of the studies used in the literature search were completely related to the topic.


Another study explored the effect that urban green spaces have on the public’s happiness. <a name="cite_ref-2"></a>[<sup>2</sup>](#cite_note-2) They used satellite imagery to analyze this relationship in over 60 different countries as well as the correlation with each country’s GDP. The result of this study was that urban green spaces have high correlations with more developed countries, those with higher GDPs. These green spaces are important as they serve as a place of community for people’s happiness. We believe that this study did a good job of exploring the relationship between urban green spaces and overall happiness. However, for our project we will not be using the correlation between mental health and GDP values.

```python
# Created a dictionary to match each response to a numerical value
answer_dict = {'Never': 0, 'Almost never': 1, 'Sometime': 2, 'Fairly often': 3, 'Very often': 4}
def stress_number(response):
    return answer_dict[response]
```

```python
df_stress_copied = df_stress.copy()
# df_stress_copied[df_stress_copied.columns[2:]] = df_stress_copied[
#     df_stress_copied.columns[2:]].apply(lambda x: stress_number(x), axis=0)
df_stress_copied = df_stress_copied.drop(columns = ['uid', 'type'])

for col in range(len(df_stress_copied.columns)):
    df_stress_copied[df_stress_copied.columns[col]] = df_stress_copied[df_stress_copied.columns[col]].apply(stress_number)

df_stress_copied
```

```python
df_stress_leftout = df_stress[['uid', 'type']]
df_stress_leftout
```

```python
df_stress_copied.insert(loc = 0, column = 'uid', value = df_stress_leftout['uid'])
df_stress_copied.insert(loc = 1, column = 'type', value = df_stress_leftout['type'])
```

# Ethics & Privacy

Biases could exist if certain demographics or groups are more or less likely to participate in health surveys. This could lead to specific groups being under or overrepresented which can affect the generalizability of the findings. In this dataset on student life in Dartmouth, only students who attend Dartmouth are included in this study so it may not necessarily include a diverse range of students. This limits the generalizability of findings to a broader population of college students. Different universities may have varying student demographics and lifestyles. In order to protect the privacy of individuals, their personal information must be anonymized or aggregated. It is important to review the terms of use for datasets because some sources may have data sharing, publication, or commercial use restrictions. This study relies on smartphone data and students who do not have a smartphone will be excluded which introduces a device bias. 

Data collection methods could introduce bias because the data on mental health is being gathered through surveys. People with a higher well being level are more likely to participate in our survey opposed to those with a lower well being level. It is important to collect data considering all population groups because this research should aim to minimize disparities. As well, the specific method that we collect data by also matters because if we collect data from individuals who have more access to physical activity, their results would always show information about high accessibility to physical activity spaces, introducing a biased data sample.


The first step would be to identify possible biases in the population, and from there we can narrow it down to which ones specifically affect our study. In this case, possible biases could negatively affect minorities such as racial or socioeconomic minorities. One group may have more data than another leading to a bias, so it will be important to include populations we believe will be a minority into our dataset. We will also make sure data is clean, so there is no missing data that can create biases. If biases are detected during or after our analysis, we will prepare by having a plan to update our question and system. If biases are found after our analysis we will acknowledge them in our conclusions and discussion.


Asking about mental health concerns is something most people tend to keep private, certain groups of people are also more susceptible to certain mental health issues that are unrelated to the amount of green space in an area. In terms of equitable impact, the analysis may be highlighting environmental disparity or taking an environmental stance. 

To combat privacy concerns, we can deal with these issues by being transparent about why we are conducting this analysis to the individuals we are observing. For potential biases, we will make sure data is clean. We will also keep data anonymous to maintain confidentiality.
